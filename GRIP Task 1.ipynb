```python
# import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
```

```python
# Reading the Data 
data = pd.read_csv('https://raw.githubusercontent.com/AdiPersonalWorks/Random/master/student_scores%20-%20student_scores.csv')
print("Import successful")
data.head(5)
```

```python
#Visualize the distribution of scores
data.plot(x='Hours', y='Scores', style='*')  
plt.title('Hours vs Percentage', size=30)  
plt.xlabel('Hours studied', size=15)  
plt.ylabel('Percentage obtained', size=15)  
plt.show()
```

### <span style='color:black'> From the above graph we can observe a positive correlation between the hours of study and the scores obtained. Plotting a regression line aids the previous statement effectively. </span>

```python
#Regression plot
sns.regplot(x= data['Hours'], y= data['Scores'])
plt.title('Regression Plot of Hours vs Percentage',size=30)
plt.xlabel('Hours studied', size=15)
plt.ylabel('Percentage obtained', size=15)
plt.show()
```

### <span style='color:black'> Now, the line clearly indicates a positve correlation. Hence, more hours spent on studying, increased score. </span>

## <span style='black'> Traning the Model </span>

```python
#Assign x and y values
x = data.iloc[:, :-1].values  
y = data.iloc[:, 1].values
```

### <span style='black'> Step 1: Split the Data </span>

```python
train_x, val_x, train_y, val_y = train_test_split(x, y, random_state = 0)
```

### <span style='black'> Step 2: Decide the Algorithm based on requirement </span>
### <span style='black'> Regression algorithms are used if there is a relationship between the input variable and the output variable. This is apt for our task. Linear regression can be adopted as we observe a linear relationship between the entities. </span>

```python
#Train the model
reg = LinearRegression()
reg.fit(train_x, train_y)
print("Model trained")
```

### <span style='black'> Step 3: Use test data to make predictions </span>
